```python
def get_Diff_Values_Colum(df_column, diff_value=[], sep=',', replace=''):
    column = df_column.dropna()
    for i in column:
        if sep not in i and i not in diff_value:
            diff_value.append(i)
        else:
            for data in map(lambda x:x.strip(), re.sub(replace, '', i).split(sep)):
                if not data in diff_value:
                    diff_value.append(data)
    return dict(enumerate(sorted(diff_value)))
```

 A. Object Column NaN Values: Reassign Group Mode Values

```python
import numpy as np
import pandas as pd

def Object_NaN_Values_Reassign_Group_Mode(df, groupby, column, inplace=True):
    # Internal function to handle the actual imputation
    def make_NaN_and_fill_mode(df, groupby, column, inplace=True):
        # 1. Standardize missing values (Fix for NumPy 2.0 np.nan)
        # Using .loc avoids the 'SettingWithCopy' warning in your original code
        df.loc[df[column].isin([None, 'nan', 'NaN', '']), column] = np.nan

        # 2. THE FIX: Replace 'stats.mode(x)[0][0]' with a safe Pandas mode
        def get_mode_safe(x):
            m = x.dropna().mode()
            # Only access [0] if a mode exists, otherwise return NaN
            return m.iloc[0] if not m.empty else np.nan

        result = df.groupby(groupby)[column].transform(lambda x: x.fillna(get_mode_safe(x)))

        if inplace:
            df[column] = result
        else:
            return result

    # Run and Print Reporting (Keeping your original structure)
    if inplace:
        # Before Assigning NaN values
        nan_count_before = df[column].isna().sum()
        if nan_count_before > 0:
            print(f'\nBefore Assigning: {column}:', f'have {nan_count_before} NaN Values', end='\n')

        a = df.groupby(groupby)[column].apply(list)
        print(f'\nBefore Assigning Example {column}:\n', *a.head().values, sep='\n', end='\n')

        # Execution
        make_NaN_and_fill_mode(df, groupby, column, inplace)

        # After Assigning NaN values
        nan_count_after = df[column].isna().sum()
        if nan_count_after > 0:
            print(f'\nAfter Assigning: {column}:', f'have {nan_count_after} NaN Values', end='\n')
        else:
            print(f'\nAfter Assigning: {column}: successfully filled all NaNs', end='\n')

        b = df.groupby(groupby)[column].apply(list)
        print(f'\nAfter Assigning Example {column}:\n', *b.head().values, sep='\n', end='\n')
    else:
        return make_NaN_and_fill_mode(df, groupby, column, inplace)
```

```python
# Define Outlier Range
def get_iqr_lower_upper(df, column, multiply=1.5):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 -q1

    lower = q1-iqr*multiply
    upper = q3+iqr*multiply
    affect = df.loc[(df[column]<lower)|(df[column]>upper)].shape
    print('Outliers:', affect)
    return lower, upper
```

```python

# Reassign Wrong Values and Show Function
def Numeric_Wrong_Values_Reassign_Group_Min_Max(df, groupby, column, inplace=True):

    # Identify Wrong values Range using Pandas mode
    def get_group_min_max(df, groupby, column):
        # We find the mode for each group, then find the overall min/max of those modes
        group_modes = df[df[column].notna()].groupby(groupby)[column].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
        return group_modes.min(), group_modes.max()

    # Assigning Wrong values
    def make_group_NaN_and_fill_mode(df, groupby, column, inplace=True):
        mini, maxi = get_group_min_max(df, groupby, column)

        # assign Values outside the "mode range" to NaN
        # Use .mask for better performance
        col_with_nans = df[column].mask((df[column] < mini) | (df[column] > maxi))

        # fill with local mode (using your original logic but safe indexing)
        mode_by_group = df.groupby(groupby)[column].transform(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
        result = col_with_nans.fillna(mode_by_group)

        if inplace:
            df[column] = result
        else:
            return result

    # --- Reporting Logic (Keeping your original structure) ---
    if inplace:
        # Before Assigning NaN values
        nan_count_before = df[column].isna().sum()
        print(f'\nBefore Assigning: {column} has {nan_count_before} NaN Values')

        print("\nExisting Min, Max Values:", df[column].agg(['min', 'max']), sep='\n')

        mini, maxi = get_group_min_max(df, groupby, column)
        print(f"\nGroupby {groupby}'s Mode range: min: {mini}, max: {maxi}")

        a = df.groupby(groupby)[column].apply(list)
        print(f'\nExample lists before:\n', *a.head().values, sep='\n')

        # Execution
        make_group_NaN_and_fill_mode(df, groupby, column, inplace)

        # After Assigning NaN values
        nan_count_after = df[column].isna().sum()
        print(f'\nAfter Assigning: {column} has {nan_count_after} NaN Values')

        b = df.groupby(groupby)[column].apply(list)
        print(f'\nAfter Assigning Example {column}:\n', *b.head().values, sep='\n')
    else:
        return make_group_NaN_and_fill_mode(df, groupby, column, inplace)
```
